**What is NORMALISING?**
Normalizing is the process that aligns data values to a common scale. This makes machine learning easier and finding the relationship between data easier and more efficient.
**Normalization Techniques:-**
1] Min-Max Scaling
2]Z Score Normalisation
3]Decimal Normalisation
4]Log Scaling
5]Robust Scaling

```python
from sklearn.preprocessing import LabelEncoder
dframe['Gender']=LabelEncoder().fit_transform(dframe['Gender'])
dframe #only the gendar column is normalised
```

```python
from sklearn.preprocessing import Normalizer #importing the normalizer
scaler=Normalizer() #initializing the normaliser
scaled_data=scaler.fit_transform(dframe) #fit and transforming the data---helps to find the maen and std dev and scale the data
scaled_data=pd.DataFrame(scaled_data,columns=dframe.columns) #convert normalized data to a new data frame which is scaled_data
scaled_data.head()
```

**MinMax Scaling**:-Tranforms features into range of 0 to 1.
Xnormalized = X – Xmin / Xmax – Xmin (here X is the value to be normalized)

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
print(scaler.fit_transform(dframe))
```

**Z Score Normalization**:-

If a value is exactly equal to the mean of all the values of the feature, it will be normalized to 0. If it is below the mean, it will be a negative number, and if it is above the mean it will be a positive number. The size of those negative and positive numbers is determined by the standard deviation of the original feature. If the unnormalized data had a large standard deviation, the normalized values will be closer to 0.

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
print(scaler.fit_transform(dframe))
```

 **Log Scaling:-**
 Log scaling computes the log of the values to compress a wide range to a narrow range. In other words, it helps convert a skewed distribution to a normal distribution/less-skewed distribution. To perform the scaling, take the log values in a column and use them as the column instead.
 x'=log(x)
 eg:-24:log(24)=1.38


```python
from sklearn.preprocessing import FunctionTransformer
transformer=FunctionTransformer()
print(transformer.fit_transform(dframe))
```

**Robust Scaling:-**
This Scaler removes the median and scales the data according to the quantile range (defaults to IQR: Interquartile Range). The IQR is the range between the 1st quartile (25th quantile) and the 3rd quartile (75th quantile).

```python
from sklearn.preprocessing import RobustScaler
scaler=RobustScaler()
print(scaler.fit_transform(dframe))
```
