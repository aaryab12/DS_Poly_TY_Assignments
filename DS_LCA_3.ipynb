**DATA VISUALISATION**


*   Line Chart
*   Axis Label
* Legend
* Line Marker
* Count Plot


---
Data visualisation is the graphical representation of data using visual elements and patterns in data..



```python

import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
import pandas as pd

wine = load_wine()
wine_df = pd.DataFrame(data=wine.data, columns=wine.feature_names)
```

```python
wine_df.head()
```

```python
y_axis=wine_df['alcohol']
plt.plot(wine_df.index,y_axis,color='pink')
plt.xlabel("Alcohol Content")
plt.title("Chart")
plt.show()
```

```python
y_axis=wine_df['alcohol']
plt.plot(wine_df.index,y_axis,color='green',linestyle='-')
plt.xlabel("Alcohol Content")
plt.title("Chart")
plt.show()
```

```python
plt.plot(wine_df['alcohol'],wine_df['malic_acid'])
plt.xlabel('Alcohol Content')
plt.ylabel('Malic Acid')
plt.show()
```

```python
plt.plot(wine_df['hue'],wine_df['alcohol'],marker='d',color='red',linestyle='--')
plt.xlabel('Hue')
plt.ylabel('Malic Acid')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=wine_df['alcohol'],
    y=wine_df['malic_acid'],
    palette='viridis',
    legend='full'
)

# Adding labels and a legend
plt.title("Wine Dataset: Alcohol vs Malic Acid")
plt.xlabel("Alcohol")
plt.ylabel("Malic Acid")
plt.legend(title="Wine Class")  # Customize the legend
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
import pandas as pd

# Load and convert the dataset into a DataFrame
wine = load_wine()
wine_df = pd.DataFrame(data=wine.data, columns=wine.feature_names)
wine_df['target'] = wine.target

plt.figure(figsize=(8, 5))
sns.countplot(data=wine_df, x='target', palette='viridis')
plt.title('Distribution of Wine Classes')
plt.xlabel('Wine Class (Target)')
plt.ylabel('Number of Samples')
plt.show()
```
